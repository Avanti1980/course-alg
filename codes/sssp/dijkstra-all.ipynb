```python
# python自带的heapq不支持decrease-key操作
# 采用第三方库heapdict https://pypi.org/project/HeapDict/
import heapdict


def init_sssp(g, s):
    dist, pred = dict(), dict()
    for v in g:
        dist[v] = float("inf") if v != s else 0
        pred[v] = None
    return dist, pred


def dijkstra_no_negative(g, s):

    dist, pred = init_sssp(g, s)  # 初始化距离和前驱
    h, in_S = heapdict.heapdict(), dict()
    for v in g:
        h[v] = float("inf") if v != s else 0
        in_S[v] = False

    iter = 0
    while h:
        iter += 1
        print(f"第{iter}轮")
        u, d = h.popitem()  # 弹出最小元
        dist[u] = d  # 此时u的值已经是最短路径
        in_S[u] = True
        print(f"({u},{d})出队，{s}到{u}的最短路径长度确定为{d}")
        for v in g[u]:  # 遍历u的所有出边(u,v)
            if not in_S[v] and h[v] > dist[u] + g[u][v]:
                print(f"放松边({u}->{v})，{s}到{v}的最短路径：{h[v]} => {dist[u] + g[u][v]}")
                h[v] = d + g[u][v]
                pred[v] = u
        print()
    return dist, pred


g = {
    "s": {"t": 10, "y": 5},  # w(s,t) = 6, w(s,y) = 5
    "t": {"x": 1, "y": 2},  # w(t,x) = 1, w(t,y) = 2
    "y": {"t": 3, "z": 2, "x": 9},  # w(y,t) = 3, w(y,z) = 2, w(y,x) = 9
    "z": {"s": 7, "x": 6},  # w(z,s) = 7, w(z,x) = 6
    "x": {"z": 4},  # w(x,z) = 4
}

dist, pred = dijkstra_no_negative(g, "s")
print(dist)
print(pred)

g = {  # dijkstra_no_negative处理带负边的图的会输出错误的结果
    "s": {"t": 6, "y": 7},  # w(s,t) = 6, w(s,y) = 7
    "t": {"x": 5, "z": -4, "y": 8},  # w(t,x) = 5, w(t,z) = -4, w(t,y) = 8
    "y": {"z": 9, "x": -3},  # w(y,z) = 9, w(y,x) = -3
    "z": {"x": 7, "s": 2},  # w(z,x) = 7, w(z,s) = 2
    "x": {"t": -2},  # w(x,t) = -2
}

dist, pred = dijkstra_no_negative(g, "s")
print(dist)
print(pred)
```

```python
# python自带的heapq不支持decrease-key操作
# 采用第三方库heapdict https://pypi.org/project/HeapDict/
import heapdict


def dijkstra(g, s):
    dist = dict()
    pred = {s: None}  # 前驱
    h = heapdict.heapdict()
    h[s] = 0  # 将源点加入二叉堆
    in_heap = dict()
    for v in g:
        dist[v] = float("inf") if v != s else 0
        in_heap[v] = False if v != s else True

    iter = 0
    while h:
        iter += 1
        print(f"第{iter}轮")
        u, d = h.popitem()  # 弹出最小元
        print(f"({u},{d})出队")
        in_heap[u] = False
        for v in g[u]:  # 遍历u的所有出边(u,v)
            suv = dist[u] + g[u][v]
            if dist[v] > suv:
                print(f"放松边({u}->{v})，{s}到{v}的最短路径：{dist[v]} => {suv}")
                if not in_heap[v]:
                    in_heap[v] = True
                    h[v] = dist[v] = suv
                    print(f"({v},{h[v]})入队")
                else:
                    print(f"更新：({v},{h[v]}) => ({v},{suv})")
                    h[v] = dist[v] = suv
                pred[v] = u
        print()
    return dist, pred


g = {  # dijkstra_no_negative处理带负边的图的会输出错误的结果
    "s": {"t": 6, "y": 7},  # w(s,t) = 6, w(s,y) = 7
    "t": {"x": 5, "z": -4, "y": 8},  # w(t,x) = 5, w(t,z) = -4, w(t,y) = 8
    "y": {"z": 9, "x": -3},  # w(y,z) = 9, w(y,x) = -3
    "z": {"x": 7, "s": 2},  # w(z,x) = 7, w(z,s) = 2
    "x": {"t": -2},  # w(x,t) = -2
}

dist, pred = dijkstra(g, "s")
print(dist)
print(pred)
```
