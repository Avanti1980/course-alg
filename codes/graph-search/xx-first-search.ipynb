广度优先搜索 BFS

```python
def bfs(g, s):
    visited = [s]  # 初始化已访问列表
    q = [s]  # 源点入队
    while q:  # 当队列非空时
        u = q.pop(0)  # 队首元素出队
        for v in g[u]:  # 对每条边(u,v)
            if v not in visited:
                visited.append(v)
                q.append(v)  # 点v入队
    return visited


g = {
    "s": ["a", "b"],            #    a     e
    "a": ["s", "c"],            #   / \   /|
    "b": ["s", "c", "d"],       #  /   \ / |
    "c": ["a", "b", "d", "e"],  # s     c  |
    "d": ["b", "c", "e"],       #  \   / \ |
    "e": ["c", "d"],            #   \ /   \|
}                               #    b-----d

visited = bfs(g, "s")
print(visited)

visited = bfs(g, "c")
print(visited)
```

深度优先搜索 DFS 递归实现

```python
def dfs_recursion(g, u):
    if u not in visited:
        visited.append(u)
        for v in g[u]:
            dfs_recursion(g, v)

g = {
    "s": ["a", "b"],            #    a     e
    "a": ["s", "c"],            #   / \   /|
    "b": ["s", "c", "d"],       #  /   \ / |
    "c": ["a", "b", "d", "e"],  # s     c  |
    "d": ["b", "c", "e"],       #  \   / \ |
    "e": ["c", "d"],            #   \ /   \|
}                               #    b-----d

visited = []
dfs_recursion(g, "s")
print(visited)
```

深度优先搜索 DFS 栈实现

```python
def dfs_stack(g, s):
    visited = []  # 初始化已访问列表
    q = [s]  # 源点入栈
    while q:  # 当栈非空时
        u = q.pop()  # 队尾元素出队
        if u not in visited:
            visited.append(u)
            for v in g[u]:  # 对每条边(u,v)
                q.append(v)  # 点v入队
    return visited


g = {
    "s": ["a", "b"],            #    a     e
    "a": ["s", "c"],            #   / \   /|
    "b": ["s", "c", "d"],       #  /   \ / |
    "c": ["a", "b", "d", "e"],  # s     c  |
    "d": ["b", "c", "e"],       #  \   / \ |
    "e": ["c", "d"],            #   \ /   \|
}                               #    b-----d

visited = dfs_stack(g, "s")
print(visited)
```
